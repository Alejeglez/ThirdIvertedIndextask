## Graphs for inverted index third task

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Librodatos.csv', sep=';')

mean_mongodb = data['MongoDB Atlas'].mean()
mean_hazelcast = data['Hazelcast'].mean()

print(f'Media de MongoDB Atlas: {mean_mongodb}')
print(f'Media de Hazelcast: {mean_hazelcast}')

ax = data.plot(kind='bar', stacked=False, x='Book Id', legend=True)

ax.set_title('Execution time of indexing for MongoDB Atlas and Hazelcast in-memory data grid')
ax.set_xlabel('Indexed Book Gutenberg id')
ax.set_ylabel('Indexing Time in ms')

ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.show()


filtered_data = data[(data['Book Id'] != 10) & (data['Book Id'] != 100)]

mean_mongodb = filtered_data['MongoDB Atlas'].mean()
mean_hazelcast = filtered_data['Hazelcast'].mean()

print(f'Media de MongoDB Atlas: {mean_mongodb}')
print(f'Media de Hazelcast: {mean_hazelcast}')
```

```python
data = pd.read_csv('Librodatosqueries.csv', sep=';')

mean_mongodb = data['MongoDB Atlas'].mean()
mean_hazelcast = data['Hazelcast'].mean()

print(f'Media de MongoDB Atlas: {mean_mongodb}')
print(f'Media de Hazelcast: {mean_hazelcast}')

ax = data.plot(kind='bar', stacked=False, x='Query Size', legend=True)

ax.set_title('Execution time of queries response for MongoDB Atlas and Hazelcast in-memory data grid')
ax.set_xlabel('Words requested in the query')
ax.set_ylabel('Response Time in ms')

ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.show()
```
